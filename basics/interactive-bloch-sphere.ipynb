**Interactive Bloch ball**

Updated version of source: https://github.com/diemilio/streams/tree/main/bloch-ball-widgets

```python
import numpy as np
from matplotlib import cm, pyplot as plt
from IPython.display import display, Math
from ipywidgets import interactive, interactive_output, HBox, VBox, Layout, FloatSlider, Label, HTMLMath, HTML, Layout, Output

from qiskit.quantum_info import DensityMatrix
from qiskit.visualization import array_to_latex
from qiskit.visualization.bloch import Bloch
from qiskit.visualization.state_visualization import _bloch_multivector_data
from time import sleep
```

```python
def density_matrix(x,y,z):
    ρ00 = 1+z
    ρ01 = x-1j*y
    ρ10 = x+1j*y
    ρ11 = 1-z
    dm_lac = array_to_latex(1/2*DensityMatrix([[ρ00,ρ01],[ρ10,ρ11]]),
                               precision=4,
                               source=True,
                               prefix='\\rho = ')
    #dm_html = HTMLMath(value = r'\(\color{black} {' + str(dm_lac[0:len(dm_lac)-0])  + '}\)')
    #display(dm_html)
    dm_html = Output()
    with dm_html:
        display(Math(r'\color{black} {' + str(dm_lac[0:len(dm_lac)-0])  + '}'))
    display(dm_html)
    sleep(0.1) #Slows down the rendering so the display isn't so glitchy
```

```python
def plot_bloch(x,y,z):
    
    _bloch = Bloch()
    _bloch.add_vectors([x,y,z])
    _bloch.render()
    _bloch.fig
```

```python
x_slide = FloatSlider(description="x:", value=1.0,min=-1.0,max=1.0,step=1.0/10)
y_slide = FloatSlider(description="y:", value=0.0,min=-1.0,max=1.0,step=1.0/10)
z_slide = FloatSlider(description="z:", value=0.0,min=-1.0,max=1.0,step=1.0/10)

figure_out = interactive_output(plot_bloch,{'x': x_slide, 'y': y_slide, 'z': z_slide})
label_out = interactive_output(density_matrix,{'x': x_slide, 'y': y_slide, 'z': z_slide})
controls = VBox([x_slide, y_slide, z_slide])
eq_label = Output()
with eq_label:
    display(Math(r"\rho = \frac{1}{2}\begin{bmatrix} 1+z & x-\imath y \\ x+\imath y & 1-z \end{bmatrix}"))
ctrl_n_label = VBox([controls,label_out])
bloch_ball = HBox([VBox([eq_label,figure_out],layout=Layout(align_items='center')),
                     ctrl_n_label])
```

**Interactive Bloch sphere**

Updated version of source: https://github.com/diemilio/streams/tree/main/bloch-ball-widgets

```python
import numpy as np
from matplotlib import cm, pyplot as plt
from IPython.display import display
from ipywidgets import interactive, interactive_output, HBox, VBox, Layout, FloatSlider, Label, Layout, Text

from qiskit.quantum_info import Statevector
from qiskit.visualization.bloch import Bloch
from qiskit.visualization.state_visualization import _bloch_multivector_data
from time import sleep
```

```python
def state_vector(θ,φ,γ):
    alpha = np.cos(θ/2)
    beta = np.exp(1j*φ)*np.sin(θ/2)
    glob = np.exp(1j*γ)
    sv_lac = Statevector(glob*np.array([alpha,beta])).draw('latex_source',prefix='|\\psi \\rangle = ',decimals=4)
    sv_html = Output()
    with sv_html:
        display(Math(r'\color{black} {' + str(sv_lac)  + '}'))
    display(sv_html)
    sleep(0.1) #Slows down the rendering so the display isn't so glitchy
```

```python
def plot_bloch(θ,φ,γ):
    
    alpha = np.cos(θ/2)
    beta = np.exp(1j*φ)*np.sin(θ/2)
    sv = Statevector([alpha,beta])
    bloch_data = _bloch_multivector_data(sv)
    
    _bloch = Bloch()
    c_ind = 20*int(γ/(2*np.pi)*10)
    c_rgb = cm.gist_rainbow(c_ind)
    _bloch.vector_color = [c_rgb]

    _bloch.add_vectors(bloch_data)
    _bloch.render()
    _bloch.fig
```

```python
θ_slide = FloatSlider(description="θ:", value=0.0,min=0.0,max=np.pi,step=np.pi/10)
φ_slide = FloatSlider(description="φ:", value=0.0,min=0.0,max=2*np.pi,step=np.pi/10)
γ_slide = FloatSlider(description="γ:", value=0.0,min=0.0,max=2*np.pi+np.pi/10,step=np.pi/10)

figure_out = interactive_output(plot_bloch,{'θ': θ_slide, 'φ': φ_slide, 'γ': γ_slide})
label_out = interactive_output(state_vector,{'θ': θ_slide, 'φ': φ_slide, 'γ': γ_slide})

controls = VBox([θ_slide, φ_slide, γ_slide])
ctrl_n_label = VBox([controls,label_out])
eq_label = Output()
with eq_label:
    display(Math(
    r'\color{black}{| \psi \rangle = e^{\imath \gamma} \left['
    r'\cos\left(\frac{\theta}{2}\right)|0\rangle + '
    r'\sin\left(\frac{\theta}{2}\right)e^{\imath\varphi}|1\rangle'
    r'\right]}'
))

bloch_sphere = HBox([VBox([eq_label,figure_out],layout=Layout(align_items='center')),
                     ctrl_n_label])
```

**Notes**
